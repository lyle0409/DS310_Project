````python
import os

def merge_json_files(output_dir, output_file):
    all_results = []

    # Lấy danh sách tất cả các file JSON trong thư mục
    for filename in os.listdir(output_dir):
        if filename.endswith(".json"):
            file_path = os.path.join(output_dir, filename)

            # Đọc dữ liệu từ file JSON
            with open(file_path, "r", encoding="utf-8") as f:
                data = json.load(f)
                all_results.extend(data)  # Thêm dữ liệu vào kết quả chung

    # Ghi kết quả vào file output
    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(all_results, f, ensure_ascii=False, indent=2)

    print(f"All files have been merged into {output_file}")

# Đường dẫn thư mục chứa các file JSON
output_dir = "/content/drive/MyDrive/DS310_NLP/Project/Output"

# Đường dẫn file JSON kết quả gộp
output_file = "/content/drive/MyDrive/DS310_NLP/Project/Output/merged_output.json"

# Gọi hàm để gộp tất cả các file
merge_json_files(output_dir, output_file)
````

````python
import ast
import re

processed_data = []
for item in data:  # Bắt đầu từ phần tử thứ 98 (chỉ mục 97)
    try:
        # Loại bỏ phần mở đầu như ```json và chuẩn hóa chuỗi
        cleaned_item = item.strip().replace("```json", "").replace("```", "")

        # Sửa các ký tự escape không hợp lệ
        cleaned_item = re.sub(r'\\(?!["/\\bfnrt])', '', cleaned_item)  # Loại bỏ escape không hợp lệ

        # Dùng json.loads() để chuyển chuỗi thành dictionary
        parsed_item = json.loads(cleaned_item)

        processed_data.append(parsed_item)
    except (json.JSONDecodeError, ValueError, SyntaxError) as e:
        print(f"Lỗi khi phân tích cú pháp JSON: {e}")
        continue

# In ra kết quả (hoặc làm gì đó với processed_data)
print(processed_data)
````
